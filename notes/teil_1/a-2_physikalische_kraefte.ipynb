\subsection{A2,a}

\begin{enumerate}
        \item Position $\vec{r}(t)$: Jedes Objekt besitzt zu einem bestimmten Zeitpunkt t genau einen festen Zustand
        \item Geschwindigkeit $\vec{v}$(t): ndert sich die Position eines Objektes in einem bestimmten Zeitbereich, so besitzt das Objekt eine Geschwindigkeit.
        \item Beschleunigung: $\vec{a}(t)$: ndert sich eine Geschwindigkeit eines Objekts in einem bestimmten Zeitraum, so wird das Objekt beschleunigt. 
        \item Kraft: $\vec{F}(t)$: Die notwendige Energie, die bentigt wird um einen Krper (Masse) zu beschleunigen, nennt man Kraft. Je schwere ein Krper, bzw, je schneller er beschleunigt werden soll, desto mehr Kraft ist aufzuwenden.
        \end{enumerate}
\subsection{A2-b}        
Vektor: $\vec{x}(y)$: Ein Vektor in der Physik gibt an, dass eine bestimmte Energie oder Kraft in eine bestimmte Richtung wirkt. So wirkt die Kraft als Beschleunigung auf ein Auto nach vorne.
